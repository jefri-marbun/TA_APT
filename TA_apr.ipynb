### 2.2 Data Transformation

```python
basket = df2.T
```

```python
basket.head()
```

```python
# Melakukan proses encoding -> Mengubah data kebentuk angka, agar sistem atau komputer dapat memahami informasi dari dataset
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.head(5)
```

3. CALCULATE

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
import pickle

with open('model_pickle', 'wb') as r:
  pickle.dump(frequent_itemsets, r)
```

```python
with open('model_pickle', 'rb') as r:
  pickle = pickle.load(r)
```

```python
rules1 = association_rules(pickle, metric="lift", min_threshold=1)
rules1.head() # 1280 rows × 9 columns
```

```python
result1 = rules1[(rules1['lift'] >= 1) & 
               (rules1['confidence'] >= 0.8)]

apr_result = result1.sort_values(by='confidence', ascending=True)
apr_result  # 38877 rows × 9 columns
```
